```python
#################################################################################################################

import os, sys

#################################################################################################################

script_path = os.getcwd()
project_path = os.path.join(script_path, '..', '..', '..')
results_dir = os.path.join(project_path, 'results', 'experiment_4')
results_path = os.path.join(results_dir, 'results_exp_4.pkl')

#################################################################################################################

sys.path.append(project_path)
from config.config_experiment_4 import N_REALIZATIONS
from src.utils.experiments_exploration_utils import process_experiment_4_results, plot_experiment_4_results

#################################################################################################################
```

```python
df, df_avg = process_experiment_4_results(results_path=results_path)
```

```python
df
```

```python
df_avg
```

```python
ref_acc = df_avg['mean_acc'].median()
std_acc = df_avg['std_acc'].median()
epsilon_acc =  std_acc + 0.1
ylim_acc = (ref_acc - epsilon_acc, ref_acc + epsilon_acc)

ref_ari = df_avg['mean_ari'].median()
std_ari = df_avg['std_ari'].median()
epsilon_ari = std_ari + 0.1
ylim_ari = (ref_ari - epsilon_ari, ref_ari + epsilon_ari)

ref_time = df_avg['mean_time'].median()
std_time = df_avg['std_time'].median()
epsilon_time = std_time + 10
ylim_time = (ref_time - epsilon_time, ref_time + epsilon_time)

print(f'ylim_acc: {ylim_acc}')
print(f'ylim_ari: {ylim_ari}')
print(f'ylim_time: {ylim_time}')
```
